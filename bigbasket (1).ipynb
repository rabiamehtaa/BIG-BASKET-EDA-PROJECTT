**STEP 6**

**Measuring discount on a certain item.**


```python
data["discount"]=data["market_price"]-data["sale_price"]
data["discount_percent"]=(data["discount"]/data["market_price"])*100

print("sample discounts")
data[["product","market_price","sale_price","discount","discount_percent"]].head()
```

**STEP 8**

**Find out the outliers from the dataset according to**
**the columns**

```python
numerical_columns=["sale_price","market_price","rating"]
for column in numerical_columns:
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lowerbound = Q1 - 1.5 * IQR
    upperbound = Q3 + 1.5 * IQR
    meanvalue = data[column].mean()
    data.loc[(data[column] < lowerbound) | (data[column] > upperbound), column] = meanvalue


```

**STEP 9**

**Create Plots or visualizations**

```python
sns.barplot(y=top.index, x=top.values,linewidth=0,alpha=1.0,color="red")
plt.ylabel("Most sold products")
plt.xlabel("Numbers")
plt.title("Sales analysis")
plt.show()
```

```python
sns.barplot(y=data["category"].value_counts()[:10].index, x=data["category"].value_counts()[:10].values,linewidth=0,alpha=1.0,color="red")
plt.title("Top And Least Sold Category")
plt.show()
```

```python

sns.distplot(data.rating, color="red")
plt.title("Rating of all products")
plt.show()
```

```python
sns.histplot(data["sale_price"],kde=True, color="red")
plt.title("Distribution of Sale Prices")
plt.xlabel("Sale Price")
plt.ylabel("Frequency")
```
